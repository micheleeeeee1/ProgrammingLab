# 5. Analisi di un dataset di video giochi
1. Quanti videogiochi sono stati pubblicati
2. Quali sono i generi più popolari? Mostralo con un bar plot
3. Plotta l'evoluzione del numero di giochi pubblicati nel tempo
4. Prova a riprodurre questo plot 



```python
# Import dei pacchetti
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caricamento del dataset
url = 'https://zenodo.org/record/5898311/files/vgsales.csv'
df = pd.read_csv(url)
df.head()
```

```python
print('sono stati pubblicati' ,df['Name'].nunique(), ' videogiochi')

generi=df['Genre'].value_counts()
generi.plot(kind='bar')
plt.title('numero di vendite per genere')
plt.xticks(rotation=45)
plt.xlabel('generi')
plt.ylabel('numero')
plt.show()



anni=df['Year'].value_counts().sort_index()
anni.plot(kind='line')
plt.title('videogiochi per anni')
plt.xlabel('anni')
plt.ylabel('numero')
plt.show()
```

```python
#sns.histplot(data= df, x= ("Genre" == 'Action'))

primi5= df.groupby('Genre')['Name'].count().sort_values(ascending=False).head(5)
venditetop5= df[df['Genre'].isin(primi5.index)].groupby('Genre')[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum()
venditetop5.plot(stacked=True, kind='bar')
plt.ylabel('miglioni di vendite')
plt.title('vendite per regione')
plt.xlabel('genere')
plt.tight_layout()
plt.show()
```
